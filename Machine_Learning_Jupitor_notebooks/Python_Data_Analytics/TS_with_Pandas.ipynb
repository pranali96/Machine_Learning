```python
def get_missing_dates(df,actual_dates,freq='B'):
    j=0
    missing_dates = []
    for i in range(len(actual_dates)):
        if actual_dates[i] != df.index[j]:
            print(actual_dates[i],df.index[j])
            missing_dates.append(actual_dates[i])
            i+=1
        else:
            j+=1
            i+=1
    return missing_dates
```

```python
def add_missing_dates(df,freq='B'):
    df.sort_index(inplace=True)
    start = df.index[0]
    end = max(df.index)
    actual_dates = pd.date_range(start,end, freq=freq)
    if ( len(actual_dates) != df.shape[0]):
        missing_dates = get_missing_dates(df,actual_dates,freq)
        df = df.append(pd.DataFrame(index=missing_dates))
        df.sort_index(inplace=True)
    return df,missing_dates
```

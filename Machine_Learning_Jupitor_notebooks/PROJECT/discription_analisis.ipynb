```python
from gingerit.gingerit import GingerIt
import pprint
import nltk




for label, row in df.iterrows():
    print(label)
    print(row['Name'])
    print(row['Short Discription '])
    text = nltk.tokenize.sent_tokenize(row['Short Discription '])
    mistakes = 0 
    for sentences in text:
        
        ginger_parser = GingerIt()

        ginger_grammar_results = ginger_parser.parse(sentences)

        pprint.pprint(ginger_grammar_results)

        ginger_corrections = ginger_grammar_results['corrections']
        mistakes = mistakes + len(ginger_corrections)
            
        pprint.pprint(ginger_corrections)
        
        
        for corrections in ginger_corrections:
            print("\t(Char #" + str(corrections['start']) + ") Use '" + corrections['correct'] + "' instead of '" + corrections['text'] + "'")
            print("*****************correct sentence will be :  " + (ginger_grammar_results['result'])+" ***")
    print("\nNumber of grammar issues found with Ginger: " + str(mistakes) + "\n")
```

```python
import pandas as pd
df = pd.read_csv("ML PROJECT DATA COLLECTION form (Responses) - Form Responses.csv")
print(df.columns) # Print column names
```

```python
# importing required modules 
import PyPDF2 
  
# creating a pdf file object 

for label, row in df.iterrows():
    #print("****** ",label)
    #print(row['Name'])
    #print(row['Resume'])
    resume=row['Resume']

    pdfFileObj = open(resume, 'rb') 
  
    # creating a pdf reader object 
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 
  
    # printing number of pages in pdf file 
    #print(pdfReader.numPages)
    num_of_pages =pdfReader.numPages
  
    # creating a page object 
    for page_number in range(num_of_pages):
        pageObj = pdfReader.getPage(page_number) 
        # extracting text from page 
        #print(pageObj.extractText())
        text=pageObj.extractText()
        print("*********  ",str(label),text)
        resume_data = zip(str(label),text)
    # closing the pdf file object 
    pdfFileObj.close() 
```

```python
print("Resume Data  ---> ",resume_data)
resume_dataset=set(resume_data)
print(resume_dataset)
#for label,resume in resume:
#    print(label,resume)
```
